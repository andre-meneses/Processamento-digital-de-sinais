# Exemplo: dizimação de um arquivo de áudio por um fator $M=2$

A aula de hoje contém um exemplo de uma dizimação de um arquivo de áudio por um fator $M=2$, implementado em *Python*.

Nesta aula, faremos uso do arquivo de audio
**581010__xcreenplay__smoking-in-the-angel-section2.wav**, obtido em https://freesound.org/browse/.

O filtro utilizado na dizimação foi projetado usando uma biblioteca de *Python* que será estudada com mais detalhes no decorrer do curso. Por enquanto, basta saber que se trata de um filtro passa-baixas (FPB) com resposta ao impulso finita (FIR) e frequência de corte $\omega_c = \pi/2$. Os coeficientes do filtro estão armazenados no arquivo **coeffs.csv**

O arquivo de áudio e o arquivo que contém os coeficientes do filtro podem ser baixados no site da turma virtual, no SIGAA.

---

O primeiro passo do código consiste na importação das bibliotecas de *Python* que serão utilizadas.

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.io import wavfile
from scipy.io.wavfile import write
from scipy.signal import welch, lfilter
```

Após importarmos as bibliotecas que serão utilizadas, carregamos o arquivo de áudio.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Carrega o arquivo
samplerate, data = wavfile.read('/content/drive/MyDrive/ColabNotebooks/DCA0118/581010__xcreenplay__smoking-in-the-angel-section2.wav')

#Carrega o arquivo em dois canais (audio estereo)
print(f"numero de canais = {data.shape[1]}")

#Tempo total = numero de amostras / fs
length = data.shape[0] / samplerate
print(f"duracao = {length}s")
```

A seguir, plotamos as figuras no domínio do tempo. (Note que estamos vizualizando as figuras com a função *plot*, que plota o sinal discreto com uma linha contínua interpolada.)

```python
#Plota as figuras ao longo do tempo

#Interpola para determinar eixo do tempo
time = np.linspace(0., length, data.shape[0])

nsampples=data.shape[0]

#Seleciona os canais esquerdo e direito
data_l = data[:, 0]
data_r = data[:, 1]

#Plota os canais esquerdo e direito
plt.figure(1)
plt.plot(time[0:nsampples], data_l, label="Canal esquerdo")
plt.legend()
plt.xlabel("Tempo [s]")
plt.ylabel("Amplitude")
plt.show()

plt.figure(2)
plt.plot(time[0:nsampples], data_r, label="Canal direito")
plt.legend()
plt.xlabel("Tempo [s]")
plt.ylabel("Amplitude")
plt.show()
```

Agora vamos carregar os coeficientes do filtro. Trata-se de um filtro passa-baixas com resposta ao impulso finita (FIR), projetado com frequencia de corte igual a $\omega_c = \pi/2$. 

```python
#Carrega os coeficientes do filtro
b = np.genfromtxt('/content/drive/MyDrive/ColabNotebooks/DCA0118/coeffs.csv', delimiter=',')

#Plota coeficientes do filtro FIR
plt.figure(3)
plt.stem(b)
plt.show()
```

Lembrando que devemos filtrar o sinal antes de descartar suas amostras, para evitar o *aliasing*. Vamos visualizar o espectro antes da filtragem.

```python
#Estima o espectro do sinal utilizando a funcao welch
x  = data_l # canal esquerdo
fs = 2
#fs = samplerate
f, Pxx_spec = welch(x, fs, 'flattop', 512, scaling='spectrum')

#Plota o espectro do sinal para frequencias normalizadas entre 0 1 pi 
#(frequencias positivas)

plt.figure(4)
plt.semilogy(f, Pxx_spec)
plt.xlabel('frequencia [rad/($\pi$)]')
plt.ylabel('Espectro')
plt.show()
```

Agora vamos realizar a operação de dizimação pelo fator $M=2$, a qual consiste em uma filtragem por um filtro passa-baixas com frequência de corte $\omega_c = \pi /M$, seguida de uma compressão pelo fator $M$. A filtragem é feita utilizando a função *lfilter* da biblioteca *ScyPy*.

```python
#Filtra os dados dos canais esquerdo e direito
filtered_data_l = lfilter(b, 1, data_l)
filtered_data_r = lfilter(b, 1, data_r)

#Downsampling dos canais esquerdo e direito
M=2

data_down_l = filtered_data_l[0:-1:M]
data_down_r = filtered_data_r[0:-1:M]

#Estima o espectro do sinal utilizando a funcao welch
x  = data_down_l # canal esquerdo
fs = 2
#fs = samplerate
f, Pxx_spec = welch(x, fs, 'flattop', 512, scaling='spectrum')

#Plota o espectro do sinal para frequencias normalizadas entre 0 1 pi 
#(frequencias positivas)

plt.figure(5)
plt.semilogy(f, Pxx_spec)
plt.xlabel('frequencia [rad/($\pi$)]')
plt.ylabel('Espectro')
plt.show()
```

Finalmente, vamos escrever criar um arquivo de áudio contendo o sinal dizimado pelo fator $M=2$.

```python
# Escrita de arquivo dizimado
audio = np.array([data_down_l, data_down_r]).T
scaled = np.int16(audio/np.max(np.abs(audio)) * 32767)
filename = '/content/drive/MyDrive/ColabNotebooks/DCA0118/signal_decimated_' + str(M) + '.wav'
write(filename, samplerate//M, scaled)
```

Vamos repetir a dizimação utilizando o mesmo filtro, porém para diferentes valores de $M$. Note que o resultado da dizimação não será correto, mas a ideia é interpretar o que acontece a partir da análise das figuras dos espectros.
